# UI/UX Usability Audit (Design Principles)

- Date: 2026-03-13
- Method: heuristic review against accessibility, hierarchy, consistency, feedback, and error prevention.
- Scope: key user flows in `lib/pages/*`, `core/*`, and app shell.

## Findings by Severity

### High
1. **Accessibility contrast and small text risk**
- `textMuted` is dark (`#666666`) on dark backgrounds and heavily used at small sizes (10-12px).
- This risks readability/contrast failures for low-vision users.

References:
- `lib/core/constants.dart:37`
- `lib/core/constants.dart:94`
- `lib/core/constants.dart:101`

Suggested fix:
- Raise muted contrast and minimum text sizes in dense contexts.

### Medium
2. **Dark-mode-only UI limits user control/flexibility**
- App is hard-pinned to dark mode.
- Violates user preference principle and may reduce readability in bright environments.

Reference:
- `lib/main.dart:114`

Suggested fix:
- Add theme preference (`system/light/dark`) in profile settings.

3. **Gesture-only clickable text/components without semantic button affordance**
- Multiple interactive elements use `GestureDetector` instead of semantic button widgets.
- This harms accessibility semantics, focus behavior, and discoverability.

References:
- `lib/pages/login_page.dart:159`
- `lib/pages/signup_page.dart:116`
- `lib/pages/responsibility_page.dart:276`

Suggested fix:
- Replace with `TextButton`/`InkWell` + semantics labels where appropriate.

4. **Inconsistent verbal tone and microcopy clarity**
- Some copy uses archaic language ("Thou appoints") while most of app is modern plain English.
- This breaks consistency and can reduce clarity for international users.

References:
- `lib/core/constants.dart:231`
- `lib/core/constants.dart:239`

Suggested fix:
- Standardize tone guide (plain, direct, international).

### Low / Fixed During This Pass
5. **Destructive action safety in Capture list**
- Swipe-delete previously had no undo.
- Now fixed with Undo snackbar action.

References:
- `lib/pages/capture_page.dart:142`
- `lib/services/providers.dart` (restore path)

6. **OTP input hint mismatch**
- OTP input used password-style hint before.
- Now fixed with an OTP-specific hint.

References:
- `lib/pages/verify_opt_page.dart:123`
- `lib/core/constants.dart` (`otpCodeHint`)

## Recommended Next UX Sprint
1. Contrast and typography pass (muted text + minimum font sizes).
2. Replace gesture-only controls with semantic button widgets.
3. Tone/copy normalization across auth, profile, and responsibility flows.
4. Add settings for theme mode and text-size sensitivity where feasible.
